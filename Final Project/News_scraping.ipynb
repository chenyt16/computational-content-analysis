```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
import time
from bs4 import BeautifulSoup
```

```python
def fox_news():
    
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Run in headless mode
    chrome_options.add_argument("--no-sandbox")  # Bypass OS security model, required for headless mode
    chrome_options.add_argument("--disable-gpu")  # Disable GPU hardware acceleration
    chrome_options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems
    driver = webdriver.Chrome(options=chrome_options)
    driver.get('https://www.foxnews.com/search-results/search?q=abortion')
    
    
    link_set = set()
    wait = WebDriverWait(driver, 20)

    # Try to click the 'Load More' button several times with waits in between
    for _ in range(5):  # Adjust the range if you need more or fewer clicks
        try:
            button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".button.load-more")))
            button.click()
            # Wait for the page to load more articles before next click
            time.sleep(5)
        except Exception as e:
            print(f"Error clicking or waiting for more articles: {e}")
            break  # Break the loop if no more 'Load More' button is found or clickable

    articles = driver.find_elements(By.CSS_SELECTOR, ".article a")
    links = [article.get_attribute('href') for article in articles if article.get_attribute('href')]
    driver.quit()
    for link in links:
        link_set.add(link)
        
    return list(link_set)
```

```python
def bbc_news():
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Run in headless mode
    chrome_options.add_argument("--no-sandbox")  # Bypass OS security model, required for headless mode
    chrome_options.add_argument("--disable-gpu")  # Disable GPU hardware acceleration
    chrome_options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems
    driver = webdriver.Chrome(options=chrome_options)
    driver.get('https://www.bbc.com/search?q=abortion')
    
    link_set = set()
    wait = WebDriverWait(driver, 10)
    # Use the CSS selector that targets the button with the specific data-testid
    for i in range(10):
        
        # Find all links within the specified class
        link_elements = driver.find_elements(By.XPATH, "//div[contains(@class, 'sc-32f23d22-1 VSDHh')]//a[@data-testid='internal-link']")
        #print(link_elements)
        for link_element in link_elements:
            link = link_element.get_attribute('href')
            #print("Found link:", link)
            link_set.add(str(link))

        button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "[data-testid='pagination-next-button']")))
        button.click()
        time.sleep(5)

        
            
    return link_set
```

```python
def cnn_news():
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Run in headless mode
    chrome_options.add_argument("--no-sandbox")  # Bypass OS security model, required for headless mode
    chrome_options.add_argument("--disable-gpu")  # Disable GPU hardware acceleration
    chrome_options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems
    driver = webdriver.Chrome(options=chrome_options)
    driver.get('https://www.cnn.com/search?q=abortion&from=0&size=10&page=1&sort=newest&types=all&section=')
    
    link_set = set()
    wait = WebDriverWait(driver, 10)
    # Use the CSS selector that targets the button with the specific data-testid
    for i in range(5):
        
        # Find all links within the specified class
        cards = driver.find_elements(By.XPATH, "//div[contains(@class, 'search__results__controls')]")

        # Extract and print the "data-open-link" attribute from each element
        open_links = [card.get_attribute("data-open-link") for card in cards]
        for link in open_links:
            link_set.add(link)

        button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "[data-testid='pagination-arrow pagination-arrow-righnt search__pagination-link text-active']")))
        button.click()
        time.sleep(5)

        
            
    return link_set
```

```python
def beck_news():
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Run in headless mode
    chrome_options.add_argument("--no-sandbox")  # Bypass OS security model, required for headless mode
    chrome_options.add_argument("--disable-gpu")  # Disable GPU hardware acceleration
    chrome_options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems
    driver = webdriver.Chrome(options=chrome_options)
    driver.get('https://www.glennbeck.com/search/?q=abortion')
    
    link_set = set()
    wait = WebDriverWait(driver, 10)
    # Use the CSS selector that targets the button with the specific data-testid
    for i in range(5):
        
        links = driver.find_elements(By.XPATH, "//a[contains(@href, '/blog/')]")#radio, contributor

        urls = [link.get_attribute('href') for link in links]
        
        for link in urls:
            link_set.add(link)

        load_more_button = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.btn.button-load-more.next-page-wrapper.action-btn")))
        driver.execute_script("arguments[0].scrollIntoView(true);", load_more_button)
        time.sleep(5)

        
            
    return link_set
```

```python
text_res = []
df = []
count = 1
for link in link_set:
    if link != 'None':
        #print(link[:-1])
        response = requests.get(link[:-1])
        #print(link)
        soup = BeautifulSoup(response.content, 'html.parser')
        #soup = soup.find_all('div', class_='article-body')
        cur_text = []
        #for body in soup:
        paragraphs = soup.find_all('p')
        for paragraph in paragraphs[1:-1]:
            cur_text.append(paragraph.get_text().replace('\n', ' '))
        text_res.append('\n'.join(cur_text))
        df.append({'File_Name': "CNNNews_{}".format(count) , 'Text_Content': ' '.join(cur_text)}, )
        count += 1
        #break
            
```
